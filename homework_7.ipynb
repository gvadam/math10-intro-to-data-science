## Task 1: Load the data and Pre-processing

1. Use `pandas` to read the csv file named `credit_risk_dataset.csv` as dataframe (The file is in the Canvas page of this homework).


2. Drop three columns, named `person_home_ownership`,`loan_intent`,`loan_grade`.


3. Drop all the samples with missing values. hint: you can use the `dropna` method, see documentation [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html).


4. For the column named `cb_person_default_on_file`, replace "Y" with integer 1, and "N" with 0. Hint: there are many ways to do this, and you can refer to one possible tutorial [here](https://www.geeksforgeeks.org/replace-the-column-contains-the-values-yes-and-no-with-true-and-false-in-python-pandas/)


5. Name your final pre-processed dataframe as `df_credit`


```python
import pandas as pd
#1. read csv file as dataframe
credit_risk = pd.read_csv('credit_risk_dataset.csv')
#2. drop three columns
credit_risk.drop(columns = ['person_home_ownership', 'loan_intent', 'loan_grade'], inplace = True)
#3. drop all samples with missing values
credit_risk.dropna(axis=0, how='any', inplace=True)
#4. replace "Y" w/1 and "N" w/0 in column 'cb_person_default_on_file'
credit_risk['cb_person_default_on_file'] = credit_risk['cb_person_default_on_file'].map({'Y':1 ,'N':0}) 
#5. final pre-processed dataframe
df_credit = credit_risk
df_credit
```

## Task 2 :  Loan Status Prediction with Logistic Regression

1. Pick up column named `loan_status`, convert it to the numpy array  `y`. Convert the dataset containing the remaining columns as numpy array `X`.


2. Create `X_train`, `X_test`, `y_train`, `y_test` by splitting the dataset. The test dataset should consist of 33% percent of the whole data. Hint: you can use the [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) function in sklearn.


3. Train a logistic regression model **with sklearn** (you can directly call the functions/methods in sklearn) on training dataset. Report the performance on test dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#1. convert column 'loan_status' to numpy array y, convert remaining columns as numpy array X
y = df_credit.loc[:,'loan_status'].to_numpy()
X = df_credit.drop(columns = 'loan_status').to_numpy()
#2. create test and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
#3. train a logistic regression model w/sklearn
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
