### Figure 1

```python
sns.displot(data=iris, x="SepalWidthCm", hue="Species", multiple="stack")
```

This distribution plot displays the number of irises (y-axis) of each species with a certain sepal width in centimeters (x-axis). Given that the graph is stacked, this allows us to compare the typical range of sepal widths for each species. The iris versicolor tends to have a shorter sepal width compared to the other irises, ranging from 2.0 cm to 3.4 cm (with 2.8 - 3.0 cm being the most common width for its species). The sepal width for iris virginica can range between 2.4 cm to 3.8 cm. Lastly, the iris setosa tends to have a larger sepal width compared to the others at 3.0 cm up to 4.4 cm (only a small number of iris setosas have a sepal width below 3.0 cm).

### Figure 2

```python
sns.kdeplot(data=iris, x="PetalLengthCm", hue="Species", multiple="stack")
```

This kernel density estimation graph displays the density (y-axis) of each iris species petal length (x-axis). For the iris setosa, the skinny tall plot means that there is a higher concentration of iris setosas with a petal length of around 1.8 cm, and it does not vary much among all of them. The iris versicolor has a slightly taller and wider plot compared to the iris virginica, indicating that there is more diversity when it comes to petal length but that most of them are commonly found with a petal length of 5 cm. The iris virginica is shifted to the right of iris versicolor, meaning that most of iris virginica have a longer petal length around 5.5 cm.

### Figure 3

```python
sns.pairplot(data=iris, hue="Species")
```

This plot displays all of the possible combinations of variables paired up together. For example, when comparing the relationship between petal length and petal width between all three species, you can see that all of the iris setosas have the shortest petal length and the shortest petal width, while iris virginica has the longest petal length and and longest petal width. These findings suggest that there is a positive correlation between petal length and petal width, and that there is a pattern in size between all three species (with setosa having an overall smaller petal, virginica having a larger petal, and versicolor being in the middle). When it comes to sepal width versus sepal length, iris setosa has the widest sepal and the shortest sepal length. Iris virginica tends to have the longest sepal length but around the same sepal width as iris versicolor.
